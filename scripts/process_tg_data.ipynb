```python
import pandas as pd
import json
import numpy as np
```

```python
FILEPATH = r"D:\trading_llm_agent\result.json"

with open(FILEPATH, "r", encoding="utf-8") as f:
    data = json.load(f)
    
```

```python
messages = data["messages"]
```

```python
messages[-2]
```

```python
from pydoc import text
import tqdm


processed_messages = []

for msg in tqdm.tqdm(messages):

    
    if msg["type"] != "message":
        continue

    message_date = msg["date"]

    message_text = ""

    for text_data in msg["text"]:
        
        if type(text_data) == str:
            message_text += text_data
            continue

        if text_data["type"] == "text_link":
            if "t.me" in text_data["href"]:
                continue
            if "Bitmar" in text_data["text"]:
                continue
            
            message_text += text_data["text"] + "(" + text_data["href"] + ")"
        else:
            message_text += text_data["text"]

    message_text = message_text.replace("\n\n", "\n")
    
    processed_messages.append({
        "message_date": message_date,
        "message_text": message_text,
        "username": msg["from"] if "from" in msg else "anonymous"
    })
```

```python
df = pd.DataFrame(processed_messages)
```
